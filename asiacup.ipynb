FIRSTLY THE ODI DATA

```python
# matches played between two teams 
all_teams = odi['Team'].unique()
mat_bet2 = pd.DataFrame(0, index=all_teams, columns=all_teams)

for index, row in odi.iterrows():
    mat_bet2.loc[row['Team'], row['Opponent']] += 1
    mat_bet2.loc[row['Opponent'], row['Team']] += 1

print(mat_bet2)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(mat_bet2, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.title('Matches Played Between Teams')
plt.xlabel('Opponent Team')
plt.ylabel('Team')
plt.show()
```

```python
win_loss = pd.DataFrame(0,index=all_teams, columns=['Wins','Losses'])

for index, row in odi.iterrows():
    if row['Result'] == 'Win':  
        win_loss.loc[row['Team'], 'Wins'] += 2
        win_loss.loc[row['Opponent'], 'Losses'] += 2
    elif row['Result'] == 'Loss':
        win_loss.loc[row['Opponent'], 'Wins'] += 2
        win_loss.loc[row['Team'], 'Losses'] += 2

# Calculate total matches played for winning percentage
win_loss['Total Matches'] = (win_loss['Wins'] + win_loss['Losses'])
win_loss['Winning Percentage'] = (win_loss['Wins'] / win_loss['Total Matches']) * 100
win_loss['Loss Percentage'] = (win_loss['Losses'] / win_loss['Total Matches']) * 100
```

```python
win_loss
```

```python
plt.figure(figsize=(10, 6))
win_loss['Total Matches'].plot(kind='bar')
plt.title('Matches Played by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Matches Played')
#plt.xticks(rotation=45)  # Rotate team names for better readability
plt.grid(axis='y')  # Add a grid for better visibility
plt.show()
```

```python

fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Bar chart for Wins and Losses
win_loss[['Wins', 'Losses']].plot(kind='bar', ax=ax[0])
ax[0].set_title('Win-Loss Record for Each Team')
ax[0].set_ylabel('Number of Matches')
ax[0].set_xlabel('Teams')
ax[0].legend(title='Match Outcome')


# Bar chart for Winning Percentage
win_loss['Winning Percentage'].plot(kind='bar', ax=ax[1], color='skyblue')
ax[1].set_title('Winning Percentage of Each Team')
ax[1].set_ylabel('Winning Percentage (%)')
ax[1].set_xlabel('Teams')

plt.tight_layout() #for making it look better
plt.show()
```

now let's see the boundary stats after the introduction of t20is in asia cup  as t20 was introduced in asia cup in 2016

```python
odi_1 = odi[(odi['Year'] >= 1984) & (odi['Year'] <= 2014)]
odi_2 = odi[(odi['Year'] >= 2014) & (odi['Year'] <= 2016)]
odi_1.reset_index(drop=True, inplace=True)
odi_2.reset_index(drop=True, inplace=True)
```

```python
#fig, ax = plt.subplot(2,2,figsize=(12, 6))

avg_fours_beforet20= odi_1['Fours'].sum()/odi_1.shape[0]
avg_fours_aftert20 = odi_2['Fours'].sum()/odi_2.shape[0]

avg_six_beforet20= odi_1['Sixes'].sum()/odi_1.shape[0]
avg_six_aftert20 = odi_2['Sixes'].sum()/odi_2.shape[0]

avg_runs_beforet20= odi_1['Run Scored'].sum()/odi_1.shape[0]
avg_runs_aftert20 = odi_2['Run Scored'].sum()/odi_2.shape[0]

avg_wicket_beforet20= odi_1['Wicket Lost'].sum()/odi_1.shape[0]
avg_wicket_aftert20 = odi_2['Wicket Lost'].sum()/odi_2.shape[0]

avg_extras_beforet20= odi_1['Extras'].sum()/odi_1.shape[0]
avg_extras_aftert20 = odi_2['Extras'].sum()/odi_2.shape[0]

avg_rr_beforet20= odi_1['Run Rate'].sum()/odi_1.shape[0]
avg_rr_aftert20 = odi_2['Run Rate'].sum()/odi_2.shape[0]

avg_sr_beforet20= odi_1['Avg Bat Strike Rate'].sum()/odi_1.shape[0]
avg_sr_aftert20 = odi_2['Avg Bat Strike Rate'].sum()/odi_2.shape[0]
```

```python
comparison_list = ['FOUR_compare', 'SIX_compare', 'WICKET_compare', 'Extras_compare', 'Runrate_compare', 'Strikerate_compare']
COMPARE = pd.DataFrame(index=comparison_list, columns=['before t20', 'after t20'])

COMPARE.loc['FOUR_compare'] = [avg_fours_beforet20, avg_fours_aftert20]
COMPARE.loc['SIX_compare'] = [avg_six_beforet20, avg_six_aftert20]
COMPARE.loc['WICKET_compare'] = [avg_wicket_beforet20, avg_wicket_aftert20]
COMPARE.loc['Extras_compare'] = [avg_extras_beforet20, avg_extras_aftert20]
COMPARE.loc['Runrate_compare'] = [avg_rr_beforet20, avg_rr_aftert20]
COMPARE.loc['Strikerate_compare'] = [avg_sr_beforet20, avg_sr_aftert20]

# Create a separate DataFrame for runs
RUN_COMPARE = pd.DataFrame(index=['RUN_compare'], columns=['before t20', 'after t20'])
RUN_COMPARE.loc['RUN_compare'] = [avg_runs_beforet20, avg_runs_aftert20]

# Create a separate DataFrame for strike rate
STRIKE_COMPARE = pd.DataFrame(index=['STRIKE_compare'], columns=['before t20', 'after t20'])
STRIKE_COMPARE.loc['STRIKE_compare'] = [avg_sr_beforet20, avg_sr_aftert20]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(25, 6))  # 1 row, 3 columns

# Plot the first set of metrics (excluding runs and strike rate)
COMPARE.plot(kind='bar', ax=axs[0])
axs[0].set_title("Comparison of Metrics Before and After T20 (Excluding Runs and Strike Rate)")
axs[0].set_ylabel("Average")
axs[0].set_xticklabels(COMPARE.index, rotation=0)  # Ensure the labels are horizontal

# Plot the runs separately
RUN_COMPARE.plot(kind='bar', ax=axs[1])
axs[1].set_title("Run Comparison Before and After T20")
axs[1].set_ylabel("Average Runs")
axs[1].set_xticklabels(RUN_COMPARE.index, rotation=0)  # Ensure the labels are horizontal

# Plot the strike rate separately
STRIKE_COMPARE.plot(kind='bar', ax=axs[2])
axs[2].set_title("Strike Rate Comparison Before and After T20")
axs[2].set_ylabel("Average Strike Rate")
axs[2].set_xticklabels(STRIKE_COMPARE.index, rotation=0)  # Ensure the labels are horizontal

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()


```
